```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Sim_new_1000000steps.txt',sep='\s+',header=None)
data = pd.DataFrame(data)
x = data[0]
y = data[2]
plt.figure(dpi=1200)
plt.plot(x, y,'bo', markersize=1)
a, b = np.polyfit(x, y, 1)
plt.xlabel("Episode")
plt.ylabel("Average reward")
plt.show()
```
