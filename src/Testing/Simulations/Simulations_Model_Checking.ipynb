```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
"""
This simulation shows the average reward per episode for the CQI algorithm and Resource gathering model.

The average results among 10 runs of simulation are displayed. Since each run had a different number of
episodes, we only plot the values up to the minimum number of episodes among those 10 runs. Only successful
runs (those for which the avg. reward in last 20 steps is at least three times biggest than the avg. reward
in the first 20 steps). In this particular run, 1 of 11 runs was unsuccessful. The parameters used are the
default parameters from our code.

Responsible function: save_full_stats()
"""

data = pd.read_csv('Sim_new_1000000steps.txt',sep='\s+',header=None)
data = pd.DataFrame(data)
x = data[0]
y = data[2]
plt.figure(dpi=1200)
plt.plot(x, y,'bo', markersize=1)
a, b = np.polyfit(x, y, 1)
plt.xlabel("Episode")
plt.ylabel("Average reward")
plt.show()
```
